1. A plaintext message of 128 bits is turned into the starting position of a 4 x 4 grid as below and the keyword also populates a second grid.

| Plaintext | | Keystring |
|-----------|-|-----------|
| $ \begin{matrix}  01 & 89 & FE & 76 \\ 23 & AB & DC & 54 \\ 45 & CD & BA & 32 \\ 67 & EF & 98 & 10 \\ \end{matrix} $ | | $ \begin{matrix}		0F & 47 & 0C & AF \\ 15 & D9 & B7 & 7F \\ 71 & E8 & AD & 67 \\ C9 & 59 & D6 & 98 \\ \end{matrix} $ | 
 

```octave
% Define a cell array with hexadecimal values
plain_text = {'01','89','FE','76' ; '23','AB','DC','54' ; '45','CD','BA','32' ; '67','EF','98','10' };
key_string = {'0F','47','0C','AF' ; '15','D9','B7','7F' ; '71','E8','AD','67' ; 'C9','59','D6','98' };
% Initialize an empty 4x4 matrix
pt_matrix = zeros(4, 4);
ks_matrix = zeros(4, 4);
% Convert each hex value to decimal and store it in the matrix
for i = 1:4
    for j = 1:4
        pt_matrix(i, j) = hex2dec(plain_text{i, j});
        ks_matrix(i, j) = hex2dec(key_string{i, j});
    end
end

% Display the matrix
disp(pt_matrix)
disp(ks_matrix)
```

a)	Use the keystring grid to perform the first Round Key operation (this involves adding the two grids cell by cell as an XOR operation).

b)	Use the S-Box to perform Byte Substitution on the 4 x 4 grid that results from a)

c) 	Perform the Shift Rows step on the 4 x 4 grid that results from b)

d) 	Perform the Mix Columns step on the 4 x 4 grid that results from c) using the matrix $ M = 
\begin{pmatrix}
2 & 3 & 1 & 1 \\
1 & 2 & 3 & 1 \\
1 & 1 & 2 & 3 \\
3 & 1 & 1 & 2 \\
\end{pmatrix} $

e) 	If the columns of the keystring grid are w0, w1, w2, w3 then construct the values for 

x1	(rotating w3 one to the left/upwards) 

y1	(the result of byte substitution on x1)

z1 	(the result of adding R1 = 01 00 00 00 to y1  )

f)	Hence find w4 = w0 + z1 		w5 = w4 + w1		w6 = w5 + w2		w7 = w6 + w3

g)	Add the columns w4  w5  w6  w7	to the grid that results from step d).
Compare your answer to the third grid in column 1 of the table on Slide 36 of this week’s lecture. They should be the same.
